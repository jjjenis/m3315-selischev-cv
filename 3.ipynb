```python
import matplotlib.pyplot as plt
import cv2
```

```python
def rectangle_numpy(img, pt1, pt2, color=(0, 255, 0), thickness=2):
    result = img.copy()

    x1, y1 = pt1
    x2, y2 = pt2

    h, w = result.shape[:2]

    x1 = max(0, min(x1, w-1))
    x2 = max(0, min(x2, w-1))
    y1 = max(0, min(y1, h-1))
    y2 = max(0, min(y2, h-1))

    for t in range(thickness):
        for x in range(x1, x2):
            result[y1+t, x] = color
            result[y2-t, x] = color

        for y in range(y1, y2):
            result[y, x1+t] = color
            result[y, x2-t] = color

    return result
```

```python
img_bgr = cv2.imread("/Users/denisandreevic/PycharmProjects/m3315-selischev-cv/white-large-square-emoji-1224280781.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
result = rectangle_numpy(img_rgb, (200, 800), (1000, 1000), color=(255, 0, 0), thickness=4)
```
